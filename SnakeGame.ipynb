```python
import pygame
import time
import random
 
pygame.init()

# Define colors of the snake
black = (0, 0, 0)
red = (255, 0, 0)
green = (0, 255, 0)
purple = (200, 100, 240)
yellow = (255, 255, 102)

# General information about the display
dis_width = 800
dis_height = 600
dis = pygame.display.set_mode((dis_width, dis_height))
pygame.display.set_caption('S N A K E   G A M E')
 
clock = pygame.time.Clock()
 
snake_block = 20
snake_speed = 5
 
font_style = pygame.font.SysFont("arial", 20)
score_font = pygame.font.SysFont("arial", 20)



# FUNCTION FOR SCORING 
def Your_score(color, score): 
    """This function displays the score of each snake - after accepting the 
    snake color and score as parameters from the calling program"""
    
    if color == "green" :
        value = score_font.render(color + "Snake Score: " + str(score), True, yellow)
        dis.blit(value, [0, 0])
    elif color == "purple" :
        value = score_font.render(color + "Snake Score: " + str(score), True, yellow)
        dis.blit(value, [0, 20])

        

def draw_snake(snake_block, snake_list, color):    
    """This function accepts the values for the snake block, list to track length of snake 
    and the snake color and draws the snake out on the screen."""
    for x in snake_list:
        pygame.draw.rect(dis, color, [x[0], x[1], snake_block, snake_block])


def message(msg, color):   
    """ This function accepts a string as the message parameter and the color of snake
    and displays the string in the defined font and style"""
    msg = font_style.render(msg, True, color)
    dis.blit(msg, [dis_width/15, dis_height/3])
 
 
def main():    
    """ This method drives the game end to end.
    The positioning of the snakes at the beginning of the game is defined here. 
    The method tracks the movement of snake through user input from keyboard 
    and updates scores of each snake on eating the fruit or removes points of the snake
    that hits the wall
    The method stops the game when either snake reaches the Max score defined and provides
    the option of either starting a new game or quitting the game altogether
    """
    game_over = False
    game_close = False
 
    green_x1 = dis_width / 2
    green_y1 = dis_height / 2
    purple_x1 = dis_width / 2
    purple_y1 = dis_height / 2
 
    green_x1_change = 0
    green_y1_change = 0
    purple_x1_change = 0
    purple_y1_change = 0
 
    green_snake_list = []
    green_length_of_snake = 1
    green_snake_score = green_length_of_snake - 1
    purple_snake_list = []
    purple_length_of_snake = 1
    purple_snake_score = purple_length_of_snake - 1
    max_score = 10
 
    green_foodx = round(random.randrange(0, dis_width - snake_block) / 20.0) * 20.0
    green_foody = round(random.randrange(0, dis_height - snake_block) / 20.0) * 20.0
    purple_foodx = round(random.randrange(0, dis_width - snake_block) / 20.0) * 20.0
    purple_foody = round(random.randrange(0, dis_height - snake_block) / 20.0) * 20.0
    
    while not game_over:
        
        while game_close == True:  
            
            # Displays the results when the game ends
            if green_snake_score >= max_score :
                message("Green snake has Max scores! Game Over! Press C-Play Again or Q-Quit", red)
            elif purple_snake_score >= max_score :
                message("Purple snake has Max scores! Game Over! Press C-Play Again or Q-Quit", red)
            else:
                message("                  It's a tie! Game Over! Press C-Play Again or Q-Quit", red)
                
            Your_score("green", green_snake_score)
            Your_score("purple", purple_snake_score)
            pygame.display.update()
 
            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    if event.key == pygame.K_c:
                        main()
                        
 
        # Processes the motion of the snakes based on the keys pressed
        # The green snake is controled by the arrow keys and the purple snake is controled by WASD 
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    green_x1_change = -snake_block
                    green_y1_change = 0
                elif event.key == pygame.K_RIGHT:
                    green_x1_change = snake_block
                    green_y1_change = 0
                elif event.key == pygame.K_UP:
                    green_y1_change -=snake_block
                    green_x1_change = 0
                elif event.key == pygame.K_DOWN:
                    green_y1_change = snake_block
                    green_x1_change = 0
                elif event.key == pygame.K_a: #left
                    purple_x1_change = -snake_block
                    purple_y1_change = 0
                elif event.key == pygame.K_d: #right
                    purple_x1_change = snake_block
                    purple_y1_change = 0
                elif event.key == pygame.K_w: #up
                    purple_y1_change -=snake_block
                    purple_x1_change = 0
                elif event.key == pygame.K_s: #down
                    purple_y1_change = snake_block
                    purple_x1_change = 0
         
        # If the snakes hit the wall, their lengths and scores reduce.   
        if green_x1 == dis_width or green_x1 == 0:
            if green_x1 == dis_width:
                green_x1_change = -snake_block
                green_y1_change = 0
            else:
                green_x1_change = snake_block
                green_y1_change = 0
            if green_snake_score > 0:
                green_snake_score -= 1
                green_length_of_snake -= 1
                del green_snake_list[-1]
            
        if green_y1 == dis_height or green_y1 == 0:
            if green_y1 == dis_height:
                green_x1_change = 0
                green_y1_change = -snake_block
            else:
                green_x1_change = 0
                green_y1_change = snake_block
            if green_snake_score > 0:
                green_snake_score -= 1
                green_length_of_snake -= 1
                del green_snake_list[-1]

        if purple_x1 == dis_width or purple_x1 == 0:
            if purple_x1 == dis_width:
                purple_x1_change = -snake_block
                purple_y1_change = 0
            else:
                purple_x1_change = snake_block
                purple_y1_change = 0
            if purple_snake_score > 0:
                purple_snake_score -= 1
                purple_length_of_snake -= 1
                del purple_snake_list[-1]

        if purple_y1 == dis_height or purple_y1 == 0:
            if purple_y1 == dis_height:
                purple_x1_change = 0
                purple_y1_change = -snake_block
            else: 
                purple_x1_change = 0
                purple_y1_change = snake_block
            if purple_snake_score > 0:
                purple_snake_score -= 1
                purple_length_of_snake -= 1
                del purple_snake_list[-1]
            
            
        # Updates the position variables of both snakes
        green_x1 += green_x1_change
        green_y1 += green_y1_change
        purple_x1 += purple_x1_change
        purple_y1 += purple_y1_change
        
        dis.fill(black)
        
        # Draws food
        pygame.draw.circle(dis, green, [green_foodx + snake_block/2, green_foody + snake_block/2], snake_block/2)
        pygame.draw.circle(dis, purple, [purple_foodx + snake_block/2, 
                                purple_foody + snake_block/2], snake_block/2)
        
        
        green_snake_Head = []
        green_snake_Head.append(green_x1)
        green_snake_Head.append(green_y1)
        purple_snake_Head = []
        purple_snake_Head.append(purple_x1)
        purple_snake_Head.append(purple_y1)
        green_snake_list.append(green_snake_Head)
        Your_score("green", green_snake_score)
        purple_snake_list.append(purple_snake_Head)
        Your_score("purple", purple_snake_score)
        
        # Ends the game if max score is reached
        if green_snake_score >= max_score or purple_snake_score >= max_score:
            game_close = True
            
        if len(green_snake_list) > green_length_of_snake:
            del green_snake_list[0]
        if len(purple_snake_list) > purple_length_of_snake:
            del purple_snake_list[0]
                
        draw_snake(snake_block, green_snake_list, green)
        draw_snake(snake_block, purple_snake_list, purple)

 
        pygame.display.update()
 
        # Checks if a snake has collected its food
        if green_x1 == green_foodx and green_y1 == green_foody:
            green_foodx = round(random.randrange(0, dis_width - snake_block) / 20.0) * 20.0
            green_foody = round(random.randrange(0, dis_height - snake_block) / 20.0) * 20.0
            green_length_of_snake += 1
            green_snake_score += 2
        if purple_x1 == purple_foodx and purple_y1 == purple_foody:
            purple_foodx = round(random.randrange(0, dis_width - snake_block) / 20.0) * 20.0
            purple_foody = round(random.randrange(0, dis_height - snake_block) / 20.0) * 20.0
            purple_length_of_snake += 1
            purple_snake_score += 2
 
        clock.tick(snake_speed)
 
    pygame.quit()
    quit()
 
 
if __name__ == "__main__":
    main()
```
